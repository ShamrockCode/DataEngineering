### Part 2: Feature Variance

Feature variance is a rather simple way of predicting if a given feature will have influence on the outcome of a trained model. The principle is that, if a feature has low variance, it cannot have much influence on the model prediction. As an extreme, if we have a variance of 0 in a feature across all samples, i.e. we have the same value for this feature in all samples, then this feature cannot be used to distinguish samples and is useless as a predictor.

However, the converse is not true. A high variance in a feature does not necessarily mean that it is a good predictor. You could imagine a feature that just contains noise with high amplitude. It might have high variance, but is meaningless. Or you can have a feature that has high variance but no influence on the outcome. For example, would hair length in cm have any influence on a baseball players salary? That said, variance can be a useful measure of the strength of a predictor.

The variance of a set of features is given by:

$$
S_{i}^{2} = \frac{\sum_{j=1}^n\left(\textbf{X}_{i,j} - \overline{\textbf{X}}_i \right)^2}{n - 1}
$$

where $\textbf{X}_{i,j}$ is the $j$'th sample of the $i$'th feature, $\overline{\textbf{X}}_i$ is the mean of all the samples of the $i$'th feature, and $n$ is the total number of samples.

Variance thresholding doesn't examine the relationship between the feature $\textbf{X}$ and the output $\textbf{y}$. This has the disadvantage that you can't test if the feature actually has an influence on the output. However, even though in this case we use it in a supervised learning context, it does mean that we can use variance thresholding for unsupervised learning.

#### Notes:

The `StandardScaler` will make it so every feature has a variance of 1 (unless the features started out with a variance of 0 to begin with) and a mean of 0. Therefore, features scaled with `StandardScaler` are useless for thresholding on variance, as there is no practical way to distinguish them.

# Task 2: Dimensionality Reduction via PCA

PCA is one of the most commonly used unsupervised transforms, and one of the most commmon means to manipulate data for machine learning. You touched on the PCA in E-tivity 2, where we investigated linear algebra. Here we will use it to reduce the numbers of features needed for a machine learning algorithm.

In the last task, we removed features. The first part, we just used some statistics on the features themselves, in independence of the other features and of the output. Then, we looked at the correlation between features and the output.

What PCA does is look at correlations *between features*. If we have high correlation between two or more features, PCA will find vectors in the feature space that best describe all features. It doesn't remove features, rather it creates a new feature space, and projects all samples to this feature space. The basis of the new feature space is a linear combination of the original features. Maybe a bit crudely, you can think of it as combining features.

### Linear Discriminant Analysis (LDA)  
  
Linear Discriminant Analysis (LDA) is a classical binary classification algorithm.Its main idea is to map all the samples onto one-dimensional axes in a way that is based on dimensionality reduction, and then set a threshold to differentiate the samples. Most commonly used as a dimensionality reduction technique in the preprocessing step of pattern classification and machine learning applications. The goal is to project the dataset onto a lower dimensional space with good class separability to avoid overfitting ("dimensionality catastrophe") and to reduce computational cost.    

Linear Discriminant Analysis (LDA) and Principal Component Analysis (PCA) are both linear transformation techniques, often used for dimensionality reduction. PCA can be described as an "unsupervised" algorithm because it "ignores" class labels and its goal is to find the directions (so-called principal components) that maximise the variance of the data set. In contrast to PCA, LDA is "supervised" and computes directions ("linear discriminants") whose axes maximise the partitioning across multiple classes.  
  
Advantages
1. Prior knowledge experience of categories can be used in the dimensionality reduction process, whereas PCA (unsupervised learning) cannot use prior knowledge of categories  
2. LDA sample classification relies on mean rather than variance, which is better than PCA algorithm
  
Disadvantages
1. LDA is not suitable for dimensionality reduction of non-Gaussian distributed samples.
2. LDA downgrading up to the number of categories N-1 dimensions, if we downgrade the dimension is greater than N-1, then we can not use LDA
3. LDA may overfit the data  
  
While it may sound intuitive that LDA outperforms PCA in multicategorisation tasks where category labels are known, this is not always the case.For example, comparisons of classification accuracies for image recognition after using either PCA or LDA show that PCA tends to outperform LDA if the number of samples per category is relatively small In practice, it is not uncommon to use both LDA and PCA: for example, PCA dimensionality reduction followed by LDA dimensionality reduction(A. M. Martinez and A. C. Kak, "PCA versus LDA," in IEEE Transactions on Pattern Analysis and Machine Intelligence, vol. 23, no. 2, pp. 228-233, Feb. 2001. https://ieeexplore.ieee.org/abstract/document/908974/citations#citations.).

So what is a "good" feature subspace?  
  
Assume that we are projecting a d-dimensional dataset into a k-dimensional space. So how do we know what size we should choose for k (k = the number of dimensions of the new feature subspace), and how do we know if we have a feature space that represents our data as "good"?
We will compute the feature vectors (components) from our dataset and will count them in a scatter matrix (i.e., inter-class scatter matrix and intra-class scatter matrix). Each eigenvector is associated with an eigenvalue that tells us about the "length" or "magnitude" of the eigenvector.
If we observe that all eigenvalues have a similar magnitude, then this may be a good indicator that our data has been projected into a "good" feature space. In another case, if some eigenvalues are much larger than others, we may be interested in keeping only the eigenvectors with the highest eigenvalues, as they contain more information about the distribution of our data. The converse is also true, eigenvalues close to 0 are less informative and we may consider removing those eigenvalues that are used to construct a new eigensubspace.
  
    
Listed below are 5 general steps for performing a linear discriminant analysis. (https://sebastianraschka.com/Articles/2014_python_lda.html#lda-in-5-steps)
1. Calculate the d-dimensional mean vectors for the different categories in the data set.
2. Calculate the scatter matrix (inter-class and intra-class scatter).
3. calculate the eigenvectors of the scatter matrix (e1,e2,.... ,ed) and corresponding eigenvalues (λ1,λ2 .... ,λd)
4. descending the eigenvalues and selecting the k eigenvectors with the largest eigenvalues to form the dimensional matrix W (each column represents an eigenvector)
5. use the d x k dimensional eigenvectors to project the original dataset into a new subspace, which can be expressed using matrix multiplication: Y = X x W, where X is an n x d dimensional matrix, n denotes the number of samples, and y is the transformation matrix, which transforms the dataset from the n x d dimensional space into the new subspace
