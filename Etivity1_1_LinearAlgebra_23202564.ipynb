## Code: a. Vector Operations

Function define

```python
import numpy as np
```

```python
def scalar_multiply(scalar, vector):

    # scalar multiplication

    result = [scalar * i for i in vector]
    return result


def vector_addition(vector1, vector2):

    # addition of two vectors

    if len(vector1) != len(vector2):
      print("Error")
      return
    else:
      result = [v1 + v2 for v1, v2 in zip(vector1, vector2)]
      return result


def vector_subtraction(vector1, vector2):

    # subtraction of two vectors

    if len(vector1) != len(vector2):
        print("Error")
        return
    else:
      result = [v1 - v2 for v1, v2 in zip(vector1, vector2)]
      return result


def dot_product(vector1, vector2):

    # dot product
    # return: float type

    if len(vector1) != len(vector2):
        print("Error")
        return
    else:
      result = sum(v1 * v2 for v1, v2 in zip(vector1, vector2))
      return result


def cross_product(vector1, vector2):

    # cross product

    x = vector1[1] * vector2[2] - vector1[2] * vector2[1]
    y = vector1[2] * vector2[0] - vector1[0] * vector2[2]
    z = vector1[0] * vector2[1] - vector1[1] * vector2[0]

    return [x, y, z]
```

## Code: b. Matrix Operations

Function define:

```python
def matrix_size(matrix):

    # calculate the size of the matrix

    return (len(matrix), len(matrix[0]))


def matrix_addition(matrix1, matrix2):

    if matrix_size(matrix1) != matrix_size(matrix2):
        print("Error")
        return
    else:
        result = [[matrix1[i][j] + matrix2[i][j] for j in range(len(matrix1[0]))] for i in range(len(matrix1))]
        return result


def matrix_subtraction(matrix1, matrix2):

    if matrix_size(matrix1) != matrix_size(matrix2):
        print("Error")
        return
    else:
        result = [[matrix1[i][j] - matrix2[i][j] for j in range(len(matrix1[0]))] for i in range(len(matrix1))]
        return result


def matrix_vector_multiplication(matrix, vector):

    if matrix_size(matrix)[1] != len(vector):
        print("Error")
        return
    else:
        result = [sum(matrix[i][j] * vector[j] for j in range(len(vector))) for i in range(len(matrix))]
        return result


def matrix_multiplication(matrix1, matrix2):
    if matrix_size(matrix1)[1] != matrix_size(matrix2)[0]:
        print("Error")
        return
    else:
        result = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(len(matrix2))) for j in range(len(matrix2[0]))] for i in range(len(matrix1))]
        return result


def determinant_2x2(matrix):
    # calculate the determinant of a 2x2 matrix
    if matrix_size(matrix) != (2, 2):
        print("Error")
        return
    else:
        return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]


def inverse_2x2(matrix):
    # calculate the inverse of a 2x2 matrix
    if matrix_size(matrix) != (2, 2):
        print("Error")
        return
    else:
        determinant = determinant_2x2(matrix)
    if determinant == 0:
        print("Error")
        return
    else:
        return [[matrix[1][1]/determinant, -matrix[0][1]/determinant],
              [-matrix[1][0]/determinant, matrix[0][0]/determinant]]
```

Text:

```python
import numpy as np

matrix1 = [[1, 2], [3, 4]]
matrix2 = [[5, 6], [7, 8]]
vector = [1, 2]

print("matrix size", matrix_size(matrix1))
print("text:", np.array(matrix1).shape)
print()

print("matrix addition", matrix_addition(matrix1, matrix2))
print("text:", np.add(matrix1, matrix2))
print()

print("matrix subtraction", matrix_subtraction(matrix1, matrix2))
print("text:", np.subtract(matrix1, matrix2))
print()

print("matrix vector multiplication", matrix_vector_multiplication(matrix1, vector))
print("text:", np.dot(matrix1, vector))
print()

print("matrix multiplication", matrix_multiplication(matrix1, matrix2))
print("text:", np.matmul(matrix1, matrix2))
print()

# Determinant and inverse of a 2x2 matrix
matrix_2x2 = [[1, 2], [3, 4]]
print("determinant2x2", determinant_2x2(matrix_2x2))
print("text:", np.linalg.det(matrix_2x2))
print()

print("inverse2x2", inverse_2x2(matrix_2x2))
print("text:", np.linalg.inv(matrix_2x2))

```
